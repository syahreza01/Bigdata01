**aaoi.us.txt**

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
import logging
import warnings

# Mengatur level logging
logging.getLogger("prophet").setLevel(logging.WARNING)
logging.getLogger("cmdstanpy").setLevel(logging.WARNING)
warnings.filterwarnings("ignore", category=FutureWarning)

# Membaca data dari file
file_name = '/content/aaoi.us.txt'
df = pd.read_csv(file_name, sep=",", header=0)
df['ds'] = pd.to_datetime(df['Date'])
df['y'] = df['Close']  # Target prediksi


# Menampilkan Jumlah data dalam sataset untuk memeriksa data
print(f"Jumlah data dalam dataset ({file_name.split('/')[-1]}): {len(df)}")

# Membuat model Prophet dengan daily seasonality diaktifkan
model = Prophet(daily_seasonality=True)
model.fit(df)

# Fungsi prediksi dan visualisasi
def make_prediction(freq_label, periods, freq_code):
    print(f"\n--- Prediksi untuk {freq_label.capitalize()} ---")
    future = model.make_future_dataframe(periods=periods, freq=freq_code)
    forecast = model.predict(future)

    # Visualisasi
    plt.figure(figsize=(10, 5))
    model.plot(forecast)
    plt.title(f"Prediksi {freq_label.capitalize()}")
    plt.xlabel("Tanggal")
    plt.ylabel("Nilai Prediksi")
    plt.grid()
    plt.tight_layout()
    plt.show()

    # Menampilkan hasil prediksi (5 baris terakhir)
    forecast_result = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(5)
    # Rename kolom untuk pemahaman yang lebih baik
    forecast_result_renamed = forecast_result.rename(columns={
        'ds': 'Tanggal [Date]',
        'yhat': 'Prediksi [Close]',
        'yhat_lower': 'Batas Bawah',
        'yhat_upper': 'Batas Atas'
    })

    print(forecast_result_renamed.to_string(index=False))

# Frekuensi prediksi
frequencies = {
    "tahun": (60, "ME"),
    "triwulan": (15, "ME"),
    "bulan": (5, "ME"),
    "minggu": (35, "D"),
    "hari": (5, "D")
}

# Menjalankan prediksi untuk setiap frekuensi
for freq_label, (periods, freq_code) in frequencies.items():
    make_prediction(freq_label, periods, freq_code)
```

```python
import pandas as pd
import pandas as lag_plot
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from pandas.plotting import lag_plot

df = pd.read_csv("/content/dgaz.us.txt")
df.head()
```

```python
print(df.head())
print(df.shape)
print(df.columns)
```

```python
df[['Close']].plot()
plt.title("Dgaz")
plt.show()
```

```python
dr = df.cumsum()
dr.plot()
plt.title('Dgaz Cumulative Returns')
```

```python
plt.figure(figsize=(10,10))
lag_plot(df['Open'], lag=5)
plt.title('Dgaz Autocorrelation plot')
```

```python
df['Date'][1445]
```

```python
train_data, test_data = df[0:int(len(df)*0.8)], df[int(len(df)*0.8):]
plt.figure(figsize=(12,7))
plt.title('Dgaz Prices')
plt.xlabel('Dates')
plt.ylabel('Prices')
plt.plot(df['Open'], 'blue', label='Training Data')
plt.plot(test_data['Open'], 'green', label='Testing Data')
plt.xticks(np.arange(0,1445, 300), df['Date'][0:1445:300])
plt.legend()
```

```python
def smape_kun(y_true, y_pred):
    return np.mean((np.abs(y_pred - y_true) * 200/ (np.abs(y_pred) + np.abs(y_true))))
```

```python
train_ar = train_data['Open'].values
test_ar = test_data['Open'].values

history = [x for x in train_ar]
print(type(history))
predictions = list()
for t in range(len(test_ar)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test_ar[t]
    history.append(obs)
    #print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test_ar, predictions)
print('Testing Mean Squared Error: %.3f' % error)
error2 = smape_kun(test_ar, predictions)
print('Symmetric mean absolute percentage error: %.3f' % error2)
```

```python
plt.figure(figsize=(12,7))
plt.plot(df['Open'], 'green', color='blue', label='Training Data')
plt.plot(test_data.index, predictions, color='green', marker='o', linestyle='dashed',
         label='Predicted Price')
plt.plot(test_data.index, test_data['Open'], color='red', label='Actual Price')
plt.title('Dgaz Prices Prediction')
plt.xlabel('Dates')
plt.ylabel('Prices')
plt.xticks(np.arange(0,1445, 300), df['Date'][0:1445:300])
plt.legend()
```
